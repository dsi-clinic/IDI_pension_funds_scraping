```python
import pdfplumber
import pandas as pd
```

```python
pdf=pdfplumber.open("sampension.pdf")


```

```python
tabs = []

for page in pdf.pages:
    tabs.append(page.extract_tables())

#tabs
```

```python
biglist=sum(tabs, [])
biggerlist=sum(biglist, [])
```

```python
table = pd.DataFrame(biggerlist[1:], columns=['Investment', 'Market Value in DKK', 'ISIN'])

print(table.head())
```

```python
table.to_csv('sampension.csv')
```
