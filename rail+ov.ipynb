```python
import pdfplumber
import pandas as pd
import re
```

```python
pdf = pdfplumber.open("rail+ov.pdf")
```

```python

tabs = []

for page in pdf.pages:
        text = page.extract_text()
        if text:
            pattern = re.compile(r'^(.+?)\s+€\s*([\d.,]+)\s+([\d.,]+%)$', re.MULTILINE)
            matches = pattern.findall(text)
            for match in matches:
                tabs.append(match)

tabs
```

```python
table = pd.DataFrame(tabs, columns=['Enterprise', 'Market Value (€)', '% of Total Portfolio'])
table
```

```python
table.to_csv("rail+ov.csv")
```
