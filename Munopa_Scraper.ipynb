```python
import pdfplumber
import csv

with pdfplumber.open("KPA.pdf") as pdf:
    with open("KPA_output.csv", "w", newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        
        # Loop through each page in the PDF, starting page indexing at 1
        for i, page in enumerate(pdf.pages, start=1):
            # Dynamically build the footer text for each page (if it varies by page)
            footer_text = f"KPA Pension | kpa.se\n{i}\n106 85 Stockholm | 0771–650 500"
            text = page.extract_text()
            
            # If text is found, remove the footer and any extra whitespace
            if text:
                text = text.replace(footer_text, "").strip()
            else:
                text = "NO TEXT FOUND"
            
            # Split the cleaned text into individual lines using splitlines()
            lines = text.splitlines()
            
            # Write each line as a separate row to the CSV file
            for line in lines:
                writer.writerow([line])
```

```python
import pdfplumber
import csv
import re

# Regex to match the line format: company name, ISIN, market value, share percent
pattern = re.compile(r'^(.*?)\s+([A-Z]{2}[A-Z0-9]{10})\s+([\d.,]+)\s+([\d.,]+\s*%)$')

with pdfplumber.open("pka.pdf") as pdf:
    with open("pka_output.csv", "w", newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow(["Company", "ISIN", "Market Value (DKK)", "Share %"])  # Header

        for i, page in enumerate(pdf.pages, start=1):
            text = page.extract_text()
            if text:
                lines = text.splitlines()

                for line in lines:
                    match = pattern.match(line.strip())
                    if match:
                        company, isin, value, percent = match.groups()
                        writer.writerow([company, isin, value, percent])
```

```python
import pdfplumber
import csv
import re

with pdfplumber.open("AP.pdf") as pdf:
    with open("AP_output.csv", "w", newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow([
            "Holding", "Domicile", "No. of Shares", "Fair Value (MSEK)",
            "ISIN", "BB Ticker", "Share Capital (%)", "Voting Capital (%)"
        ])

        for page in pdf.pages:
            text = page.extract_text()
            if not text:
                continue

            lines = text.splitlines()

            for line in lines:
                # Skip section headers and irrelevant lines
                if "Shares and participations" in line or "Holding" in line or "TOTAL" in line:
                    continue

                # Split the line into parts
                parts = line.strip().split()

                # Attempt to match lines with the expected structure
                if len(parts) >= 10:
                    try:
                        # Extract from the end of the line
                        voting_capital = parts[-1]
                        share_capital = parts[-2]
                        bb_ticker = ' '.join(parts[-4:-2])
                        isin = parts[-5]
                        fair_value = parts[-6]
                        no_of_shares = ' '.join(parts[-9:-6])
                        domicile = parts[-10]
                        holding = ' '.join(parts[:-10])

                        writer.writerow([
                            holding, domicile, no_of_shares, fair_value,
                            isin, bb_ticker, share_capital, voting_capital
                        ])
                    except Exception as e:
                        print(f"Line skipped due to error: {line}\n{e}")


```

```python
import pdfplumber
import csv
import re

# Output file
output_csv = "ap3_fixed_income_output.csv"

# Regex pattern: matches each line in the format
pattern = re.compile(
    r"^([A-Z]{3})\s+([A-Za-z&/\- ]+?)\s+(.+?)\s+([\d\s]+)\s+(\d{1,2}/\d{1,2}/\d{4})\s+([A-Z0-9]{12})$"
)

with pdfplumber.open("ap3fixedincomeholdings.pdf") as pdf:
    with open(output_csv, "w", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        writer.writerow(["Currency", "Type", "Name", "Market Value (TSEK)", "Maturity Date", "ISIN"])

        for page in pdf.pages:
            text = page.extract_text()
            if not text:
                continue

            lines = text.splitlines()
            for line in lines:
                line = line.strip()
                match = pattern.match(line)
                if match:
                    currency, asset_type, name, value, maturity_date, isin = match.groups()
                    writer.writerow([
                        currency,
                        asset_type.strip(),
                        name.strip(),
                        value.replace(" ", ""),  # remove spaces from large numbers
                        maturity_date,
                        isin
                    ])
```

```python
import pdfplumber
import csv

with pdfplumber.open("ap3fixedincomeholdings.pdf") as pdf:
    with open("ap3_fixed_income_output.csv", "w", newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow([
            "Currency", "Type", "Name", "Market Value (TSEK)", 
            "Maturity Date", "ISIN"
        ])

        for page in pdf.pages:
            text = page.extract_text()
            if not text:
                continue

            lines = text.splitlines()
            for line in lines:
                line = line.strip()

                # Skip header lines or empty lines
                if not line or "Currency Type Name" in line or "Holdings" in line or line == "#":
                    continue

                parts = line.split()

                if len(parts) >= 6:
                    currency = parts[0]
                    asset_type = parts[1]
                    isin = parts[-1]
                    maturity_date = parts[-2]
                    market_value = parts[-3]
                    name = ' '.join(parts[2:-3])

                    writer.writerow([
                        currency,
                        asset_type,
                        name,
                        market_value.replace(",", ""),  # clean comma formatting
                        maturity_date,
                        isin
                    ])
```

```python
import pdfplumber
import csv

with pdfplumber.open("ap3privateequity.pdf") as pdf:
    with open("ap3_private_equity_output.csv", "w", newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow([
            "Fund/Private Equity Firm", "Currency", "Commitment (in millions)", "Vintage Year"
        ])

        for page in pdf.pages:
            text = page.extract_text()
            if not text:
                continue

            lines = text.splitlines()
            for line in lines:
                line = line.strip()

                # Skip headers and irrelevant lines
                if (
                    not line or 
                    "Private equity investments" in line or 
                    "AP3's commitments" in line or 
                    "Totalt Vintage" in line or
                    "Fund/Private equity firm" in line or 
                    "as of" in line.lower() or 
                    "All amounts" in line or
                    line == "#"
                ):
                    continue

                parts = line.split()

                if len(parts) >= 4:
                    vintage_year = parts[-1]
                    commitment = parts[-2]
                    currency = parts[-3]
                    fund_name = ' '.join(parts[:-3])

                    writer.writerow([
                        fund_name,
                        currency,
                        commitment,
                        vintage_year
                    ])
```

```python
import pdfplumber
import csv

with pdfplumber.open("ap3foreignequityholdings.pdf") as pdf:
    with open("ap3_foreign_equity_output.csv", "w", newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow([
            "Currency", "Instrument Group", "Company Name", "Units",
            "Value (TSEK)", "% Ownership", "% Voting Power", "ISIN"
        ])

        for page in pdf.pages:
            text = page.extract_text()
            if not text:
                continue

            lines = text.splitlines()
            for line in lines:
                line = line.strip()

                # Skip headers, footers, and page tags
                if (
                    not line or
                    "Currency Instrument group" in line or
                    "ownership" in line.lower() or
                    "Holdnings as of" in line or
                    line.startswith("#") or
                    line.startswith("AP3") or
                    line.startswith("Foreign Equity Holdings")
                ):
                    continue

                parts = line.split()
                if len(parts) < 8:
                    continue  # skip broken lines

                # Extract from known positions
                currency = parts[0]
                instrument_group = parts[1]
                isin = parts[-1]
                voting_power = parts[-2]
                ownership = parts[-3]
                value = parts[-4]
                units = parts[-5]
                company_name = " ".join(parts[2:-5])

                writer.writerow([
                    currency,
                    instrument_group,
                    company_name,
                    units.replace(",", ""),
                    value.replace(",", ""),
                    ownership,
                    voting_power,
                    isin
                ])
```

```python
import pdfplumber
import pandas as pd
import re

# Path to the PDF file
pdf_path = "ap1holdings.pdf"

# This list will store extracted data
data = []

# Open the PDF and read each page
with pdfplumber.open(pdf_path) as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        if text:
            # Regex to capture bond data in the format: Name, Market Value, No of Shares, ISIN
            matches = re.findall(
                r'Bond\s+(.*?)\s+([\d.,]+)\s+([\d.,]+)\s+([A-Z0-9]+)',
                text
            )
            data.extend(matches)

# Convert list of tuples into a DataFrame
df = pd.DataFrame(data, columns=["Name", "Market Value (KSEK)", "No of Shares", "ISIN"])

# Clean the numerical columns
df["Market Value (KSEK)"] = df["Market Value (KSEK)"].str.replace(",", "").astype(float)
df["No of Shares"] = df["No of Shares"].str.replace(",", "").astype(float)

# Save the cleaned data to CSV
df.to_csv("ap1holdings_extracted.csv", index=False)
```

```python
import pdfplumber
import pandas as pd
import re

# Step 1: Define the path to the PDF
pdf_path_bpl = "BPL.pdf"  # make sure this matches your file name and location

# Step 2: Create a list to store the extracted rows
bpl_data = []

# Step 3: Open the PDF and loop through each page
with pdfplumber.open(pdf_path_bpl) as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        if text:
            # Step 4: Use regex to extract Issuer, Percentage, and Market Value
            matches = re.findall(
                r'([A-Z0-9a-z\s/().,&\'\-+]+?)\s+([\d.,]+)%\s+([\d.,]+)',
                text
            )
            bpl_data.extend(matches)

# Step 5: Turn into DataFrame
bpl_df = pd.DataFrame(bpl_data, columns=["Issuer", "Percentage", "Market Value (€1,000)"])

# Step 6: Clean numeric columns
bpl_df["Percentage"] = bpl_df["Percentage"].str.replace(",", ".").astype(float)
bpl_df["Market Value (€1,000)"] = (
    bpl_df["Market Value (€1,000)"]
    .str.replace(".", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)

# Step 7: Save as CSV
bpl_df.to_csv("BPL_extracted.csv", index=False)
print("Saved to BPL_extracted.csv")
```
