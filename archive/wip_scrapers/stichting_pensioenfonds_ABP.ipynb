```python
import pandas as pd
import pdfplumber
```

```python
pdf = pdfplumber.open("stichting_pensioenfonds_ABP.pdf")
```

```python
tabs = []

for page in pdf.pages:
    tabs.append(
        page.extract_tables(
            table_settings={
                "vertical_strategy": "lines",
                "join_y_tolerance": 50,
            }
        )
    )

# tabs
```

```python
tabs2 = sum(tabs, [])
tabs3 = sum(tabs2, [])
# tabs3
```

```python
for line in tabs3:
    if line == ["Regio / Land / Uitgever", "Waarde"] or line == ["", ""]:
        tabs3.remove(line)

# tabs3
```
