```python
import pdfplumber
import pandas as pd
```

```python
pdf = pdfplumber.open("danicapension.pdf")
```

```python
tabs = []



for page in pdf.pages:
    tabs.append(page.extract_tables(table_settings={"vertical_strategy": "text", "horizontal_strategy": "text", "text_keep_blank_chars": False}))


#tabs
```

```python
biglist=sum(tabs, [])
```

```python
biggerlist=sum(biglist, [])
```

```python
biggestlist=[]

for i in range(len(biggerlist)):
    if biggerlist[i]!= ['', '', '', '']:
        if biggerlist[i]!=['', '', '', '', '']:
            biggestlist.append(biggerlist[i])

#biggestlist
```

```python
test = biggestlist[300][0:-2]

test2 = " ".join(test)
#test2
```

```python
for i in range(len(biggestlist)):
    combo = []
    combo = biggestlist[i][0:-2]
    del biggestlist[i][0:-2]
    combo = "".join(combo)
    biggestlist[i].append(combo)

#biggestlist
```

```python
table = pd.DataFrame(biggestlist[4:-5], columns=['marketvalue', 'share%', 'listedshares'])
table.head()
```

```python
reorder = ['listedshares', 'marketvalue', 'share%']
table2 = table[reorder]
table2.head()
```
