```python
import pandas as pd
import pdfplumber
```

```python
pdf = pdfplumber.open("vervoer.pdf")
```

```python
tabs = []

for page in pdf.pages:
    text = page.extract_text(layout=True, x_density=4)
    tabs.append(text.strip() + "\n")

# tabs
```

```python
tabs2 = " ".join(tabs)

# tabs2
```

```python
tabs3 = tabs2.splitlines()
tabs4 = []
for tab in tabs3:
    tab = tab.strip()
    tabs4.append(tab.split("  "))

tabs4 = sum(tabs4, [])
```

```python
for tab in tabs4:
    if tab == "" or tab == " " or tab == "  " or tab == "   ":
        tabs4.remove(tab)

for tab in tabs4:
    if tab == "Overzicht beleggingen per 31-12-2024":
        tabs4.remove(tab)

for tab in tabs4:
    if "Pagina" in tab:
        tabs4.remove(tab)

for tab in tabs4:
    if "Marktwaarde" in tab:
        tabs4.remove(tab)

tabs4 = tabs4[13:]
```

```python
tabs5 = []
for tab in tabs4:
    if tab != "":
        tabs5.append(tab)

# tabs5
```

```python
del tabs5[19:23]
```

```python
chunks = [tabs5[i : i + 4] for i in range(0, len(tabs5), 4)]
chunks = chunks[:1494]

# print(chunks)
```

```python
table = pd.DataFrame(chunks, columns=["1", "2", "3", "4"])
table.head()
```

```python
shares = table[["1", "2"]].copy()
shares.columns = ["issuer name", "marketvalue x €1000"]
shares.head()
```

```python
fixed_income_securities = table[["3", "4"]].copy()
fixed_income_securities.columns = ["issuer name", "marketvalue x €1000"]
fixed_income_securities.head()
```

```python
shares.to_csv("vervoershares.csv")
fixed_income_securities.to_csv("vervoerfixedincome.csv")
```
