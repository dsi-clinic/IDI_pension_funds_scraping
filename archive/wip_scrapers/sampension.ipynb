```python
import re
```

```python
# entry_pattern = re.compile("(?P<issuer>[A-Za-z\d ]+)\n\n(?P<value>[\d\.]+)\n\n(?P<ISIN>[A-Z\d]+)") #findall
date_pattern = re.compile(r"\d+\.\d+\.\d{4}")  # search

issuer_pattern = re.compile(r"(?P<issuer>[A-Za-z\d\-\. /&()'_:,]+)")
value_isin_pattern = re.compile(
    r"(?P<value>\d[\d\. ]+) (?P<isin>[A-Z\d ]{12,})"
)

currency = "DKK"
```

```python
a = "18.573.777 DEQOOA1EWWWO"
re.search(value_isin_pattern, a)
```

```python
string_cheese = ""
issuer_col = []
values_col = []
isin_col = []


for page_num, page in enumerate(pdf.pages):
    string_cheese = ""
    try:
        im = pdf.pages[page_num].to_image(resolution=144).save("sampension.png")
        string_cheese = pytesseract.image_to_string(
            Image.open("sampension.png")
        )
        print(page_num)
    except:
        print("fail")
        # log fail
        pass

    if string_cheese:
        string_cheese = re.sub(r"’|£|\|", "", string_cheese)

        issuers = re.findall(issuer_pattern, string_cheese)
        for m in issuers:
            is_issuer = re.findall(r"\d", m)
            is_issuer = len(is_issuer)

            if is_issuer <= 2:
                issuer_col.append(m)

        values_isin = re.findall(value_isin_pattern, string_cheese)

        for m in values_isin:
            value, isin = m

            if re.search(r"\.", value):
                values_col.append(value)
                isin_col.append(isin)
            # print(m)
```

```python
issuer_col_fixed = []
for i in issuer_col:
    if re.search("[Ss]ampension", i):
        pass
    elif re.search("DKK", i):
        pass
    else:
        issuer_col_fixed.append(i)
```

```python
print(len(issuer_col_fixed))
print(len(isin_col))
```

```python
if len(issuer_col_fixed) != len(isin_col):
    last = re.search("[A-Z ]+", string_cheese)
    issuer_col_fixed.append(last.group())
```
