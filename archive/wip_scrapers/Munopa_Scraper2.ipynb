```python
import csv

import pdfplumber

# Open the PDF
amf = pdfplumber.open("amf.pdf")

# Initialize a list to collect all company names across pages
all_companies = []

# Loop through pages 1 to 11 (Python index 0 to 10)
for i in range(11):  # this includes page 0 to 10
    page = amf.pages[i]

    # Use different crop top depending on the page
    top_crop = 275 if i == 0 else 130
    pager = page.crop((50, top_crop, 595.276, 800))

    # Extract text with layout info
    tabs = pager.extract_text(layout=True)
    if not tabs:
        continue  # Skip if no text extracted

    lines = tabs.splitlines()

    # Find company names (not all-uppercase strings)
    for line in lines:
        for thing in line.split("  "):
            if thing.strip() and thing.upper() != thing:
                all_companies.append(thing.strip())

# Print a sample of the company names found
print("\nExtracted Companies:")
for company in all_companies[:20]:
    print(company)

# OPTIONAL: Save to a CSV
with open("extracted_companies.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["Company"])
    for company in all_companies:
        writer.writerow([company])

print("\n✅ Extraction complete. Saved to 'extracted_companies.csv'.")
```
