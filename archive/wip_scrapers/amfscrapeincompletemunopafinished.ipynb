```python
pager = page.crop((50, 130, 595.276, 800))
```

```python
tabs = pager.extract_text(layout=True)
```

```python
lines = tabs.splitlines()
lines
```

```python
companies = []
```

```python
for line in lines:
    for thing in line.split("  "):
        if thing != "":
            if thing.upper() != thing:
                companies.append(thing.strip())
```

```python
look_for_caps = re.compile(r"[A-Z ]{10,}")
```

```python
lines = tabs.splitlines()
```

```python
for line in lines:
    matches = look_for_caps.search(line)
    if matches:
        print(matches)
```

```python
for i in range(0, len(lines)):
    matches = look_for_caps.search(lines[i])
    if matches:
        print(f"match! at index {i}")
```
