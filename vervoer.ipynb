```python
import pdfplumber
import re
import pandas as pd
```

```python
pdf = pdfplumber.open("vervoer.pdf")
```

```python
tabs = []

for page in pdf.pages:
    text = page.extract_text(layout = True, x_density=5)
    tabs.append(text.strip())

tabs
```

```python
tabs2 = " ".join(tabs)
type(tabs2)
```

```python
pattern = re.compile(r"^(.+?)\s{2,}([\d\s.]+)\s{2,}(.+?)\s{2,}([\d\s.]+)$")
matches = pattern.findall(tabs2)
```
