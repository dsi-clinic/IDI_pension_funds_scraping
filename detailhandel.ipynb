```python
import pdfplumber
import pandas as pd
import re
```

```python
pdf = pdfplumber.open("detailhandel.pdf")
```

```python
tabs = []

for page in pdf.pages:
    tabs.append(page.extract_text())

tabs
```

```python
tabs2 = " ".join(tabs)
```

```python
pattern = re.compile(r'(.+?)\s+([\d,]+)\s+(.+?)\s+([\d,]+)')
matches = pattern.findall(tabs2)
```

```python
matches
```

```python
for item in matches:
    if item == ('Beleggingen Pensioenfonds Detailhandel per', '31', 'december', '2024'):
        matches.remove(item)
```

```python
# Organize the extracted data
data = []
for name1, value1, name2, value2 in matches:
    data.append({"Name": name1.strip(), "MarketValue(EUR)": value1.replace(",", "")})
    data.append({"Name": name2.strip(), "MarketValue(EUR)": value2.replace(",", "")})
```

```python
# Convert to DataFrame
df = pd.DataFrame(data[:-2])
print(df)
```

```python
df.to_csv("detailhandel.csv")
```
