```python
import pdfplumber
import re
import pandas as pd
```

```python
pdf = pdfplumber.open("bpfbouw.pdf")
```

```python
tabs = []

for page in pdf.pages:
    text = page.extract_text()
    if text:
            pattern = re.compile(r'^(.+?)\s+([\d.]+,\d+)$', re.MULTILINE)
            matches = pattern.findall(text)
            tabs.append(matches)

tabs
```

```python
tabs2=sum(tabs, [])
```

```python
table = pd.DataFrame(tabs2, columns=["share", "amount x €1000"])
table
```
