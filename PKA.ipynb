```python
import pdfplumber
import pandas as pd
```

```python
pdf = pdfplumber.open("PKA.pdf")
```

```python
tabs=[]
for page in pdf.pages:
    tabs.append(page.extract_tables(table_settings={"vertical_strategy": "text", "text_x_tolerance": 10}))

tabs
```

```python
tabs2 = sum(tabs, [])
tabs3 = sum(tabs2, [])
tabs3
```

```python
tabs3[1][3:]
```

```python
for i in range(len(tabs3)):
    combo = []
    combo = tabs3[i][0:-4]
    del tabs3[i][0:-4]
    combo = "".join(combo)
    tabs3[i].append(combo)

tabs3
```

```python
print(tabs3[0][3])
tabs3[0][2]
```

```python
for i in range(len(tabs3)):
    tabs3[i].append("".join(tabs3[i][2:4]))

tabs3
```

```python
table = pd.DataFrame(tabs3, columns = ["ISIN", "MV(DKK)", "no", "no2", "listed shares", "share"])
table
```

```python
table = table.drop(['no', 'no2'], axis=1)
```

```python
neworder=['listed shares', 'ISIN', 'MV(DKK)', 'share']
table = table[neworder]
table
```
